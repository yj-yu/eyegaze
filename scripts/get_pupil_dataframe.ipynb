```python
def get_pad(paths,length=16):
    paths = sorted(paths)
    if len(paths) >= length:
        offset = random.randint(0,len(paths)-16)
        return paths[offset:length+offset]
    else:
        return [paths[i%length] for i in range(length)]
```

```python
train_rows = []
valid_rows = []
test_rows = []

rows = []
header = ['key', 'rawdata', 'path', 'label']
rows.append(header)
train_rows.append(header)
valid_rows.append(header)
video_frm = glob.glob(VAS_path+'vid_frm/'+'*')
total = 0
for vf in video_frm:
    key = vf.split('/')[-1]
    mat_path = vf.replace('/vid_frm/','/gazemap_cowork/') + '.mat'
 
    row1 = [key, mat_path, vf]
    if int(key[3]) in [1,2,3,4,5]:
        train_rows.append(row1)    
    if int(key[3]) in [6]:
        valid_rows.append(row1)
        
```
